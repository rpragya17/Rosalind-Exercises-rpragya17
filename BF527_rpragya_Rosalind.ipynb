<b>Counting DNA Nucleotides<b>

```python
def run(s) :
    #initialize variables to store count of nucleotides 
    a_count = 0
    g_count = 0
    c_count = 0
    t_count = 0
    
    #for loop to loop through each character in sequence
    #add 1 to count variable corresponding to each nucleotide if present in sequence 
    for i in s:
        if i=='A': 
            a_count += 1
        elif i=='G':
            g_count += 1
        elif i=='C':
            c_count += 1
        else:
            t_count += 1
    
    #print result 
    print( a_count, c_count, g_count, t_count)
```

```python
f= open('rosalind_dna.txt', 'r') #open file containing dataset 
seq = '' #intialize variable to store sequence in 

for line in f: #parse through each line in file 
    seq+= line.rstrip() #strip line and append to empty variable 
    
run(seq) #run function defined above with data from dataset 
 
f.close() #close file containing dataset 
```

<b>Transcribing DNA into RNA<b>

```python
def transcribe(t):
    u = '' #initialise empty variable to store transcribed sequence 
    no_change_bases = ['G', 'A', 'C'] #list of bases that are present in RNA, do not undergo chang after transcription 
    for i in t:
        #if element in DNA sequence is present in list defined above, do not change in transcribed sequence
        #if element is not present in list (T), change to U in resulting sequence 
        if i in no_change_bases:
            u += i
        else:
            u += 'U'
            
    #print output 
    print(u)
```

```python
f2 = open('rosalind_rna.txt', 'r') #open file containing dataset 
t = '' #intialize variable to store dna sequence in

for line in f2: #parse through each line in file 
    t += line.rstrip() #strip line and append to empty variable 
    
transcribe(t) #run function defined above with data from dataset

f2.close()  #close file containing dataset 
```

<b>Complementing a Strand of DNA<b>

```python
def revcomp(d):
    comp= '' #intialize empty variable to store complementary sequence in 
    comp_base = {'A':'T', 'T':'A','G':'C', 'C':'G'} #define dictionary containing complementary bases 
    
    #for loop to parse through each element in sequence passed 
    #for each element, append complementary base as defined in dictionary to output sequence  
    for i in d:
        if i in comp_base:
            comp += comp_base[i]
            
    #print reversed complementary sequence as required 
    print(comp[::-1])
```

```python
f3 = open('rosalind_revc.txt', 'r') #open file containing dataset
d = '' #intialize variable to store dna sequence in

for line in f3: #parse through each line in file 
    d += line.rstrip() #strip line and append to empty variable
    
revcomp(d) #run function defined above with data from dataset

f3.close() #close file containing dataset 
```

<b>Rabbits and Recurrence Relations<b>

```python
def fib(n, k): 
    
    if n== 0:
        k = 0
    elif n == 1:
        k = 1
    else:
        k = (fib(n-1, k) + k*fib(n-2, k))    
    return k
```

```python
f10 = open('rosalind_fib.txt', 'r') #open file containing dataset
t = [] #initialise empty list to store values of n and k 

for line in f10: #parse through each line in file 
    l = line.strip() #strip line and save to variable l
    t += l.split(' ') #split line on space and save to list t 

k = int(t[1]) #define variables for function 
n = int(t[0])

print(fib(n, k)) #run function defined above with data from dataset

f10.close()  #close file containing dataset 
```

<b>Counting Point Mutations<b>

```python
def countpm(s, t):
    count = 0 #intialize counter 
    for i in range(len(s)): 
        if s[i] != t[i]: #if element in string s is not the same as that number element in string t add 1 to counter 
            count += 1
        
    print(count) #print counter 
```

<b>Calculating Protein Mass<b>

```python
def prot_mass(P):
    w = 0 #intialise variable to store protein mass 
    p_mass = {'A':71.03711,'C':103.00919,'D':115.02694,'E':129.04259,'F':147.06841,'G':57.02146,'H':137.05891,'I':113.08406,
'K':128.09496,'L':113.08406,'M':131.04049,'N':114.04293,'P':97.05276,'Q':128.05858,'R':156.10111,'S':87.03203,
'T':101.04768,'V':99.06841,'W':186.07931,'Y':163.06333} #dictionary to store mass of each amino acid 
    for i in P: 
        w += p_mass[i] #add mass corresponding to each amino acid in the sequence to total mass of protein
        
    print(round(w, 3)) #print total mass of protein, round off to 3 points after decimal
```

<b>Enumerating Gene Orders<b>

```python
import math 
import itertools as it

def permutation(n):
    print(math.factorial(n)) #print factorial of n
    
    a = it.permutations(range(1,n+1)) #generates permutations of numbers from 1 to nth element 
    for i, j in enumerate(a): #parses through each list in a 
        b = '' 
        for l in j: #parses through each element in each list in a 
            b+= str(l)+ ' ' #append value to empty string and add empty space after each value 
        print(b) #print output 
```

```python
with open("rosalind_perm.txt","r") as f7: #open file containing dataset
    line=f7.readline() #read each line in file 
    n = int(line) 
    
permutation(n) #run function defined above with data from dataset

f7.close()  #close file containing dataset 
```

<b>Computing GC Content<b>

```python
f9 = open('rosalind_gc.txt','r') #open file containing dataset
seq_dict = {} #initialise empty dictionary to store sequence ID and sequence 

for line in f9: #loop through each line in file  
    if line[0] == '>': #if line starts with '>' symbol, strip line and add to dictionary key 
        
        seqID = line.split('>')[1] #extract element at index 1 and add to variable seqID 
        seqID = seqID.rstrip() #strip line to remove newline character 
        seq_dict[seqID] = ''
    
    #sequence lines do not start with '>' symbol
    else:
        #append sequence to values of dictionary 
        seq_dict[seqID]+= line.strip()
        
f9.close()  #close file containing dataset   
```

```python
#initialize variables and counters 
gc = 0 #variable to store gc count 
seq_list = [] #list to store only sequences 
maxp = 0 #variable to store maximum gc percentage 

for val in seq_dict.values():
    seq_list.append(val) #append values of dictionary to seq_list 

for seq in seq_list: #loop through each sequence seq_list 
    for j in seq: #loop through each element in sequence 
        if j=="G" or j=="C": #add 1 to gc counter if sequence has bases G or C
            gc+=1 
            
    percent=gc/(len(seq))*100 
    gc=0 #reset value of gc counter 
    
    #if gc percentage is greater than maxp, update value of maxp otherwise make no changes 
    #save sequence ID of sequence with high gc content 
    #reset value of gc percentage 
    if percent>maxp: 
        maxp=percent
        percent=0
        seq_ID=[k for k, s in seq_dict.items() if s == seq]
        seq_ID=''.join(seq_ID) 
    else:
        maxp=maxp
        percent=0

print(seq_ID)
print(maxp) #print output
```

<b>Mendel's First Law<b>

```python
def mlaw(k, m, n):
    a = k + m + n #total population 
    
    #probability of picking two organisms that give birth to an offspring with at least one dominant allele calculated as shown in attached figure  
    prob = (4*(k*(k-1) +2*k*m +2*k*n +m*n) + 3*m*(m-1))/(4*a*(a-1)) 
    print(prob)
```

```python
f8 = open('rosalind_iprb.txt', 'r') #open file containing dataset
r = [] #initialise empty list to store values of dataset 

for line in f8: #loop through each line in file 
    l = line.strip() #strip line and append to empty variable
    r += l.split(' ') #split on blank space 

k = int(r[0]) #define variables for function 
m = int(r[1])
n = int(r[2])

mlaw(k, m, n) #run function defined above with data from dataset

f8.close()  #close file containing dataset 
```
